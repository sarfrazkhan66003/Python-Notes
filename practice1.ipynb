```python
#Given an array arr of integers. Find whether three numbers are such that the sum of two elements equals the third element.
class solution:  # class to find triplet
    def findTriplet(self,arr):   # function to find triplet
        arr.sort()        # sort the array
        n = len(arr)      # length of array
        for k in range(n-1,1,-1):    # iterate from n-1 to 1
            i,j=0 ,k-1  # initialize i and j
            while(i<j):    # iterate from i to j
                if(arr[i]+arr[j]==arr[k]):    # check if sum of two elements is equal to third element
                    return True   # if yes return true
                elif(arr[i]+arr[j]<arr[k]):   # if sum of two elements is less than third element
                    i+=1   # increment i
                else:
                    j-=1  # decrement j
        return False  #
    
arr = [1, 2, 3, 4, 5]
a = solution()    # initialize object
print(a.findTriplet(arr))
```

```python
'''Given an array of strings arr. Return the longest common prefix among each and every strings present in the array.
If there's no prefix common in all the strings, return "-1".'''
class Solution:  # class to find longest common prefix
    def longestCommonPrefix(self, arr):  # function to find longest common prefix
        if not arr: # if array is empty
            return "-1"    # return -1
        prefix = arr[0]   # initialize prefix
        for s in arr[1:]:  # iterate through the array
            while not s.startswith(prefix):  # check if prefix is common
                prefix = prefix[:-1]  # remove last character
                if not prefix:   # if prefix is empty
                    return "-1"  # return -1
        return prefix  # return prefix

arr = ["geeksforgeeks", "geeks", "geek", "geezer"]  # given array
solution = Solution()   # initialize object
print(solution.longestCommonPrefix(arr))  # print longest common prefix
```

```python
#Given a binary tree, find its height.
class solution:
    def height(self, root):
        if not root:
            return 0
        return max(self.height(root.left), self.height(root.right)) + 1
root = [1,2,3,4,5]
a = solution()
print(a.height(root))
```
