# Remove Elements from a set.

```python
'''A particular item can be removed from set usin method, discard() and remove().'''
s = {1,2,3,5,4}
print(s)

s.discard(4) #4 is removed from set s.
print(s)
```

```python
#Discard an element not present in a set s. but the value are same available
s = {1,2,3}
s.discard(5)
print(s)
```

```python
#remove an element not present in a set. there are gives a key error.
s = {1,2,3}
s.remove(5)
print(s)
```

```python
'''We can remove item using pop() method '''
s = {1,2,3,4,5,6}
s.pop() # remove  random elememt
print(s)
```

# Set Operation

```python
s1 = {1,2,3,4,5}
s2 = {3,4,5,6,7}
#union of 2 sets using | or operator
print(s1|s2)
```

```python
#another way to getting union of 2 sets
print(s1.union(s2))
```

```python
#intersection of 2 sets using & operator
print(s1 & s2)
```

```python
#use intersection Function
print(s1.intersection(s2))
```

```python
#Set difference: Set of elements that are only in set1 but not in set2.
print(s1 - s2)
```

```python
#use difference function
print(s1.difference(s2))
```

```python
'''Symmetric Diference: set of elements in both set1 and set2
#except those that are Common in both.'''
#use ^ operator
s1 = {1,2,3,4,5}
s2 = {3,4,5,6,7}
print(s1^s2)
```

```python
#use symmetric function
print(s1.symmetric_difference(s2))
```

```python
#Find issubset()
x={"a","b","c","d","e"}
y={"c","d"}
print("set 'x' is subset of 'y' ?",x.issubset(y)) #check x is subset of y.
print("set 'y' is subset of 'x' ? ", y.issubset(x)) #check y is subset of x.
```
