```python
temp = 1
even_sum = 0
odd_sum = 0
while temp<20:
      if temp%2==0:
            even_sum+=temp
            temp+=1
      else:
            odd_sum+=temp
            temp+=1
print("Even Sum ",even_sum,"Odd sum",odd_sum)
```

```python
#Even sum and odd sum
n = int(input("Enter a number"))
odd_sum=0
even_sum=0
temp = 1
while temp<=n:
      if temp%2==0:
            even_sum+=temp
            temp+=1
      else:
            odd_sum+=temp
            temp+=1
print("Even Sum ",even_sum,"Odd sum",odd_sum)
```

```python
#Factorial
n = int(input("Enter a number"))
fact=1
ans=1
while fact<=n:
      ans*=fact
      fact+=1
print(ans)
```
