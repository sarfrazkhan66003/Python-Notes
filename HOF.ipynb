```python
def return_fizz_buzz_cnt(lst):
    five_cnt=0
    three_cnt=0
    seven_cnt=0
    for i in lst :
        if i%7==0 :
            seven_cnt+=1
        elif i%3==0:
            three_cnt+=1
        elif i%5==0:
            five_cnt+=1
    return (five_cnt,three_cnt,seven_cnt)

```

```python
def hof_cnt(fun,lst):
    result=0
    for i in lst:
        if fun(i):
            result+=1
    return result
lst=[3,5,7,12,15,30,21,1,21,56,65]
x=lambda X:X%3==0
y=lambda Y:Y%5==0
z=lambda Z:Z%7==0

print(hof_cnt(x,lst))
print(hof_cnt(y,lst))
print(hof_cnt(z,lst))
```
