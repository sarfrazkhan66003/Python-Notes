```python

class Node:
    def __init__(self, value):
        # value, left, right
        self.val = value
        self.left = None
        self.right = None
        
```

```python
# Example of Binary tree

root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
```

```python
def display(root):
    if root is None:
        return None
    print(root.val, end= " ")
    display(root.left)
    display(root.right)

display(root)
```

```python
# Sum of nodes of binary tree

def sum_of_nodes(root):
    if root is None:
        return 0
    return root.val + sum_of_nodes(root.left) + sum_of_nodes(root.right)
```

```python
sum_of_nodes(root)
```

```python
def max_of_tree(root):

    if (root is None)  or (root.left is None and root.right is None):
        return root
    
    max_left = max_of_tree(root.left)
    right_max = max_of_tree(root.right)
    
    return max( max_left, right_max, root,
               key  = lambda x: x.val if x else float('-inf'))
```

```python
max_of_tree(root).val
```

```python
# height of a binary tree

def find_height(root):
    if root is None:
        return -1
    return 1 + max(find_height(root.left), find_height(root.right))

find_height(root)
```

```python
# size of tree
def size_of_tree(root):
    if root is None:
        return 0
    return size_of_tree(root.left) + size_of_tree(root.right) + 1


size_of_tree(root)
```

```python
# Preorder traversal

def preorder(root):
    if root is None:
        return
    print(root.val, end = " ")
    preorder(root.left)
    preorder(root.right)
```

```python
preorder(root)
```

```python
# Inorder traversal

def inorder(root):
    if root is None:
        return
    
    inorder(root.left)
    print(root.val, end = " ")
    inorder(root.right)
```

```python
inorder(root)
```

```python
# Postorder traversal

def postorder(root):
    if root is None:
        return
    
    postorder(root.left)
    postorder(root.right)
    print(root.val, end = " ")
```

```python
postorder(root)
```

# Breadth First Traversal

```python
from collections import deque

def level_order(root):
    if root is None:
        return
    queue = deque([root])

    while queue:
        node = queue.popleft()
        print(node.val, end= " ")
        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)

```

```python
level_order(root)
```

### Print nth level

```python
from collections import deque

def print_nth_level(root, n):
    if root is None:
        return
    
    level = 0
    queue = deque([root])

    while queue:
        level_size = len(queue)
        if level == n:
            for i in range(level_size):
                node = queue.popleft()
                print(node.val, end = " ")
            return
        
        for _ in range(level_size):
            node = queue.popleft()
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)

        level+=1

    print(" Level not found ")

        
```

```python
print_nth_level(root, n=1)
```

### Leet code 100 - same tree

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

from typing import Optional

class Solution:
    def isSameTree(self, p: Optional[TreeNode], q: Optional[TreeNode]) -> bool:
        if not p and not q:
            return True
        if not p or not q:
            return False
        if p.val != q.val :
            return False
        return self.isSameTree(p.left, q.left)  and self.isSameTree(p.right, q.right)
        
```

### Invert a binary tree LC 226

```python
class Solution:
    def invertTree(self, root: Optional[TreeNode]) -> Optional[TreeNode]:
        if not root:
            return None
        
        # Swapped the children
        tmp = root.left
        root.left = root.right
        root.right = tmp

        # Recursively invert the subtrees
        self.invertTree(root.left)
        self.invertTree(root.right)

        return root
```
