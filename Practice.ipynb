```python
# 3. Write a program to find the factorial of a number
n = int(input('enter your number'))
def cal_fact(n):
      fact = 1
      for i in range(1,n+1):
            fact = fact * i
      return fact
print(cal_fact(n))
      
```

```python
# 5. Write a program to print even in string and odd number in string
def print_even_odd(start,end):
      print('even numbers: ')
      for i in range(start,end+1):
            if i%2 == 0:
                  print(i,end=' ')
                  print('\n')
                  
      print('odd number: ')
      for i in range(start,end+1):
            if i%2 != 0:
                  print(i,end=' ')
                  print('\n')
                  
start = int(input('enter start value: '))
end = int(input('enter end value: '))

print_even_odd(start,end)
      
```

```python
# 6. Write a recursive function to calculate the sum of 1st n natural numbers.
def sum_natural(n):
      if n == 1:
            return 1
      else:
            return n + sum_natural(n-1)

n = int(input('enter number: '))

print(sum_natural(n))
```

```python
# 7. Write a recursive function to print all elements in a list.
def print_lst(lst):
      if len(lst) == 0:
            return
      print(lst[0])
      print_lst(lst[1:])
      
lst = [1,2,3,4,5]
print_lst(lst)
```

```python
def print_l(lst,idx=0):
      if(idx==len(lst)):
            return
      print(lst[idx])
      print_l(lst,idx+1)
      
fruit = ['apple','mango','banana']
print_l(fruit)
      
```

```python
''' 8. PALINDROME_NUM_
Given an integer x, return true if x is palindrome integer. 
An integer is a palindrome when it reads the same backward as forward. For example, 
121 is a palindrome while 123 is not.

Note : Always print your output using print function.

Input
The input will be an integer

Output
The output should be in True or False

Example
Input: 121

Output: True'''

def palindrome_num(n):
      return str(n) == str(n)[::-1]

n = int(input('enter number: '))

print(palindrome_num(n))
```
