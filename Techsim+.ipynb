```python
'''Count all letters, digits, and special symbols from a 
given string
str1 = "P@#yn26at^&i5ve"
output will be 
Total counts of chars, digits, and symbols 
Chars = 8 
Digits = 3 
Symbol = 4'''

str1 = "P@#yn26at^&i5ve"
count1 = 0
count2 = 0
count3 = 0
for i in str1:
      if i.isalpha():
            count1+=1
      elif i.isdigit():
            count2+=1
      else:
            count3+=1
print(f'Total counts of chars, digits, and symbols \nChars = {count1} \nDigits = {count2} \nSymbol = {count3}')
```

```python
#Inheritance
class A:
      def __init__(self):
            print('A init')
      def method1(self): 
            print('Method 1 from Class A')
            
      def  method3(self):
            print('method 3 from class A')
class B(A):
      def __init__(self):
            print('B init')
            super().__init__()
            
      def method2(self):
            print('Method 2 from Class B')
      def method1(self):
            print('Function Overriding')
            
class C(B):
      def __init__(self):
            print('C init')
      def method1(self):
            print('method 1 from c class')
            super().method1()
            A.method1(self)
            

# obj = B()
# obj.method2()
# obj.method1()

obj1 = C()
obj1.method1()
```

```python
#Multiple Inheritance
class A:
      def __init__(self):
            print('A init')
      def method1(self): 
            print('Method 1 from Class A')
            
      def  method3(self):
            print('method 3 from class A')
class B:
      def __init__(self):
            print('B init')
            super().__init__()
            
      def method2(self):
            print('Method 2 from Class B')
      def method1(self):
            print('Function Overriding')
class D(A,B):                      
      def __init__(self):
            print('D init')
      def method_d(self):
            A.method1(self)
            B.method1(self)

d = D()
#d.method1()
d.method_d()
```
