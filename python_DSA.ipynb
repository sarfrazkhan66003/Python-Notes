```python
#509. Fibonacci Number

class Solution:
    def rec(self,n,dp):
        #base case
        if n==0 or n==1:
            return n
        
        #already calculated
        if dp[n]!=-1:
            return dp[n]
        
        #recursive case
        dp[n]=(self.rec(n-1,dp)+self.rec(n-2,dp)) #top down approach
        
        return dp[n]
    
n = 6 # Example input
dp=[-1]*(n+1) #memoization array
obj=Solution()  
print(obj.rec(n,dp)) 

        
```

```python
class Solution:
    def fib(self, n:int) -> int:
        dp=[0]*(n+1)
        dp[0]=0
        if n>0:
            dp[1]=1
        for i in range(2,n+1):
            dp[i]=dp[i-1]+dp[i-2] #bottom up approach
        return dp[n]
    
n = 6 # Example input
obj=Solution()      
print(obj.fib(n))       
```

```python
# 231. Power of Two

class Solution:
    def isPowerofTwo(self,n:int) -> bool:
        while n%2==0:
            n//=2
        return n==1
    
n = 16 # Example input
obj=Solution()
print(obj.isPowerofTwo(n))
```

```python
# with recursion
class Solution:
    def isPowerofTwo(self,n:int) -> bool:
        #base case
        if n<=0:
            return False
        if n==1:
            return True
        if n%2!=0:
            return False
        
        #recursive case
        return self.isPowerofTwo(n//2)
    
n = 16 # Example input
obj=Solution()
print(obj.isPowerofTwo(n))
```

```python
# 1137. N-th Tribonacci Number

class Solution:
    def tribonacci(self,n:int) -> int:
        dp=[0]*(n+1) #memoization array
        dp[0]=0 #base case
        if n>0: 
            dp[1]=1 
        if n>1:
            dp[2]=1 
        for i in range(3,n+1):
            dp[i]=dp[i-1]+dp[i-2]+dp[i-3] #bottom up approach
        return dp[n]
    
n = 6 # Example input
obj=Solution()
print(obj.tribonacci(n))
```

```python
# with recusion

class Solution:
    def rec(self,n,dp): 
        #base case
        if n==0:
            return 0
        if n==1 or n==2:
            return 1
        
        #already calculated
        if dp[n]!=-1:
            return dp[n]
        
        #recursive case
        dp[n]=(self.rec(n-1,dp)+self.rec(n-2,dp)+self.rec(n-3,dp)) #top down approach
        
        return dp[n]
    
    def tribonacci(self,n:int) -> int:
        dp=[-1]*(n+1) #memoization array
        return self.rec(n,dp)
    
n = 6 # Example input
obj=Solution()
print(obj.tribonacci(n))
```

```python
# 50. Power(x,n)

class Solution:
    def myPow(self,x:float,n:int) -> float:
        if n==0:
            return 1
        if n<0:
            x=1/x
            n=-n
        if n%2==0:
            return self.myPow(x*x,n//2)
        else:
            return x*self.myPow(x*x,n//2)
        
x = -3
n = 5 # Example input      
obj=Solution()      
print(obj.myPow(x,n))
```

```python
class Solution:
    def findPow(self,x,n):
        if n==0:
            return 1
        a = self.findPow(x,n//2)
        if n%2==0:
            return a*a
        else:
            return a*a*x
        
    def myPow(self,x:float,n:int) -> float:
        if n>=0:
            return self.findPow(x,n)
        else:
            return 1/self.findPow(x,n*(-1))
        
x = -3
n = 5 # Example input
obj=Solution()
print(obj.myPow(x,n))
```

```python
# 912. sort an array


from typing import List 
class Solution:
    def sortArray(self, nums: List[int]) -> List[int]:
        return sorted(nums) #using built-in sort function
    
nums = [5,2,3,1] # Example input
obj=Solution()
print(obj.sortArray(nums))
```

```python
from typing import List 
class Solution:
    def sortArray(self, nums: List[int]) -> List[int]:
        n = len(nums)
        for i in range(n):
            for j in range(n-i-1):
                if nums[j]>nums[j+1]:
                    temp = nums[j]
                    nums[j]=nums[j+1]
                    nums[j+1]=temp
                    
        return nums
    
nums = [5,2,3,1] # Example input
obj=Solution()
print(obj.sortArray(nums))
```
