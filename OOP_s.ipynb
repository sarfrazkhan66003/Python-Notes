```python
class Car:
    def __init__(self,brand,model,year):
        print(self)
        self.brand = brand
        self.model = model
        self.year = year
        
    def start_engine(self):
        print("The car engine has started.")
        
    def stop_engine(self):
        print("The car engine has stopped.")
        
my_car = Car("Toyota","Corolla",2022)
print(my_car)

your_car = Car("Honda","Civic",2023)
print(my_car.brand)
print(your_car.model)
```

```python
#Area Of circle 

class Circle:
      def __init__(self,radius):
            self.radius = radius
            
      def calculate_area(self):
            return 3.14 * self.radius * self.radius
      
circle = Circle(5)
area = circle.calculate_area()
print(area)
```

```python
#Rectangle

class Rectangle:
    width = 0
    height = 0
    def __init__(self, width, height):
        self.width = width
        self.height = height
    @classmethod  #
    def create_square(cls, side_length):
        return cls(side_length, side_length)


# Create a rectangle object using the class constructor
rectangle = Rectangle(4, 6)
print(rectangle.width, rectangle.height)  # Output: 4 6

# Create a square object using the class method
square = Rectangle.create_square(5)
#square1=rectangle.create_square(9)
print(square.width, square .height)  # Output: 5 5
```

```python
class MathUnits:
      @staticmethod
      def add(a,b):
            return a + b

result = MathUnits.add(3,5)
print(result)
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def get_name(self):
        return self.name

    def get_age(self):
        return self.age

    def increment_age(self):
        self.age += 1

if __name__ == "__main__":
    person1 = Person("Vikram", 19)
    print(f"Name: {person1.get_name()}, Age: {person1.get_age()}")

    # Increment the age
    person1.increment_age()
    print(f"Updated Age: {person1.get_age()}")
```
