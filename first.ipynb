5. Write a program that determines the eligibility for a discount based on the total
purchase amount. If the purchase amount is greater than or equal to 100, the
customer is eligible for a 10% discount. Write a program that takes the total
purchase amount as input and displays the discounted amount and the final
amount to be paid.

Input:
Enter the total purchase amount: 120

Expected Output:
Congratulations! You are eligible for a discount of 10%.

Discounted Amount: 12.0
Final Amount to be paid: 108.0

```python
amount = float(input("Enter your ammount"))

if(amount>=100):
      print("You are eligible for 10% discount")
      discount = amount*0.1
      print("Discount amount : ", discount)
      after_discount = amount - discount
      print("Now you are eligible to pay", after_discount)
      
else:
      print("Not eligible")      
```

```python
num = int(input("Enter a number"))
factorial = 1

if num<0:
      print("No Factorial for negative number")
elif num == 0:
      print("The Factorial of 0 is 1")
      
else:
      for i in range(1, num+1):
            factorial *=i
      print("Factorial of ",num ,"is",factorial)
```
