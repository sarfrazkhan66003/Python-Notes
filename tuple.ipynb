```python
# You are given a list of tuples, each containing a student's name and their exam scores. 
# Write a Python function to sort the list of tuples in descending order of scores and, in case of a tie, in ascending order of names.
# student_scores = [("Alice", 85), ("Bob", 92), ("Eve", 77), ("David", 92), ("Charlie", 88), ("Frank", 77)]
# output : [('Bob'', 92), ('David'', 92), ('Charlie', 88), ('Alice', 85), ('Frank', 77), ('Eve', 77)]

lst = [("Alice", 85), ("Bob", 92), ("Eve", 77), ("David", 92), ("Charlie", 88), ("Frank", 77)]
n = len(lst)
for i in range(n):
    for j in range(0, n - i - 1):
        if lst[j][1] < lst[j + 1][1]:
            lst[j], lst[j + 1] = lst[j + 1], lst[j]
print(lst)
```

```python
# You are given a list of tuples, each representing a person's name, age, and height. Write a Python function to filter the list and return a new tuple containing the names of people who are older than 30 and taller than 160 cm. 
# input : people_info = [("Alice", 28, 165), ("Bob", 35, 172), ("Eve", 29, 160), ("David", 45, 175), ("Charlie", 31, 158)]
# outputt : ('Bob', 'David')

lst =  [("Alice", 28, 165), ("Bob", 35, 172), ("Eve", 29, 160), ("David", 45, 175), ("Charlie", 31, 158)]
lst1 = []
for i in lst:
      if i[1]>30 and i[2]>160:
            lst1.append(i[0])
print(tuple(lst1))
```
