# Factorial of a number

```python
def factorial(n):

    # base condition
    if n==0:
        return 1
    else:
        return n*factorial(n-1)
    
print(factorial(6))
```

# Fibonacci Series

```python
def fibonacci(n):   # f(n) = f(n-1) +f(n-2)

    # base condition
    if n<=1:
        return n
    
    else:
        return fibonacci(n-1) +fibonacci(n-2)
    

fibonacci(6)
```

# Sum of Array

```python
def sum_of(arr, n):

    #base conditon
    if n == 0:
        return 0
    
    else:
        return sum_of(arr, n-1) + arr[n-1]
    

arr = [1,2,3,4,15]
print(sum_of(arr,len(arr)))
```

# print 1 to n   

```python
def print_numbers(n):
    # base condition
    if n>0:
        print_numbers(n-1)
        print(n)


print_numbers(6)
```

# Print n to 1

```python
def print_numbers(n):
    # base condition
    if n>0:
        print(n)
        print_numbers(n-1)
        

print_numbers(6)
```

# Number of ways to climb stairs


```python
def climb_stairs(n):

    if n==0:
        return 1
    elif n<0:
        return 0
    else:
        return climb_stairs(n-1)+ climb_stairs(n-2)
    
n = 2
climb_stairs(n)   # climb(1)  + climb(0)
```

# binary search

```python
def binary_search(arr, left, right, target):

    if left>right:
        return -1  # base condition
    
    mid = left + (right - left)//2

    if arr[mid] == target:
        return mid    # target is present
    
    elif arr[mid] > target:             # [12,14,18,20]
        return binary_search(arr, left, mid-1, target)
    else:
        return binary_search(arr, mid+1, right, target)

    
arr = [2,3,4,6,10,30]
target = 10

print(binary_search(arr, 0, len(arr)-1 , target))
```

# Subset of string

```python
def generate_subset(s, current = "", index =0):

    if index == len(s):
        print(current)
        print("--------------------")
        return 
    
    generate_subset(s, current + s[index], index +1)
    generate_subset(s, current , index +1)


ex = "abc"
generate_subset(ex)
```
