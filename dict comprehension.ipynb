### DICT COMPREHENSION

```python
lst1=[1,2,3,4,5]
lst2 = [i*10 for i in lst1 if i%2!=0]
print(lst2)
```

```python
d={'a':1,'b':2,'c':3,'d':4}
d2={}
for k,v in d.items():
    d2[k]=v*100
print(d2)
```

```python
d={'a':1,'b':2,'c':3,'d':4}
d2={k:v for k,v in d.items() if v>2}
print(d2)
```

```python
d={'a':1,'b':2,'c':3,'d':4,'e':5}
d2={k+'c':v*2 for k,v in d.items() if v>2}
print(d2)
```

```python
d={2:[1,2,5,6],3:[3,5,2,1],4:[3,1,2,7]}
d2={}
for k,v in d.items():
    d2[k]=v[:k]
print(d2)
```

```python
d={2:[1,2,5,6],3:[3,5,2,1],4:[3,1,2,7]}
d2={}
for k,v in d.items():
    for i in v:
        if i%2!=0:
            v*=i
            d2[k]=v
print(d2)
```
