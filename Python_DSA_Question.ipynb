### 1.Selection Sort

```python
def selection_sort(arr):
    #checking the size of an array
    n = len(arr)
    for i in range(n):
        #assume the current element is the smallest
        min_index = i #index of the smallest element is 0
        #find the smallest element index in the remaining unsorted part
        for j in range(i+1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        #swap the found smallest element with the first element
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr

arr = [64,25,172,22,101]
#1st pass i=0 min_index=0, array ke list index se n-1==min_index = 3
#[22,25,172,64,101]
#2nd pass i=1 array me 2nd se n-1
#[22,25,172,64,101]
#3rd pass i=2 min_index = 2 array ke 3 se n-1, min_index = 3
#[22,25,64,172,101]
#4th pass i=3 min_index = 3 array ke 4 se n-1, min_index = 4
#[22,25,64,101,172]
sorted_arr = selection_sort(arr)
print("Sorted array is:", sorted_arr)
```

### 2.Insertion Sort

```python
def insertion_sort(arr):
    #traverse through 1 to len(arr)
    for i in range(1, len(arr)):
        key = arr[i]
        #j is initialized to the index of the last element in the sorted portion of the array
        j = i-1
        #compare key with elements in the sorted portion on array
        while j >= 0 and key < arr[j]:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key
    return arr

arr = [12,-11,136,5,6]
#1st pass i=1 key= -11, j=0
#[-11,12,136,5,6]
#2nd pass i=2 key=136, j=1
#[-11,12,136,5,6]
#3rd pass i=3 key=5, j=2
#[-11,5,12,136,6]
#4th pass i=4 key=6, j=3
#[-11,5,6,12,136]
sorted_arr = insertion_sort(arr)
print("Sorted array is:", sorted_arr)
```

### 3.Bubble Sort

```python
def bubble_sort(arr):
    n=len(arr)
    #traverse through all elements in the array
    for i in range(n-1):
        for j in range(n-i-1):
            #swap if the element found is greater than the next element
            arr[j], arr[j+1] = arr[j+1], arr[j] 

arr = [5,1,3,4,2]
print(arr)
bubble_sort(arr)
print(arr) 
```

```python
# Question:
'''How much maximum swaps are needes to sort array of length 6?'''
def bubble_swap_count(arr):
    n=len(arr)
    count=0
    for i in range(n-1):
        is_sorted = True #flag to check if the array is already sorted
        for j in range(n-i-1):
            if (arr[j]>arr[j+1]):
                is_sorted = False #if the array is not sorted
                count += 1
                #left is more than right, swap
                arr[j], arr[j+1] = arr[j+1], arr[j]
        if is_sorted:
            break
    return count

arr = [5,4,3,2,1]
print("Maximum swaps needed to sort the array are:", bubble_swap_count(arr))
```

```python
'''Reverse Pairs (Leetcode Problem) : Given an integer array nums, return the number of reverse pairs in the 
array.
                              A reverse pair is a pair (i, j) where:
                              0 <= i < j < nums.length and
                              nums[i] > 2 * nums[j].'''
class Solution:
    
    def reversePairs(self, nums: List[int]) -> int:
        self.count = 0
        def merge(left_arr, right_arr):
            # Step 1: Count the index satisfied condition first
            i, j = 0, 0
            # note that i is index of left_arr, j is index of right_arr .
            while i < len(left_arr) and j < len(right_arr):
                if left_arr[i] > 2 * right_arr[j]:
                    self.count += len(left_arr) - i
                    j += 1
                else:
                    i += 1
            # Step 2: Merge sort left and right
            l, r = 0, 0
            res = []
            while l < len(left_arr) and r < len(right_arr):
                if left_arr[l] < right_arr[r]:
                    res.append(left_arr[l])
                    l += 1
                else:
                    res.append(right_arr[r])
                    r += 1
            return res + left_arr[l:] + right_arr[r:]
        def divide(arr):
            if len(arr) <= 1: return arr
            # divide the input array into 2 parts
            mid = len(arr)//2
            left_arr = divide(arr[:mid])
            right_arr = divide(arr[mid:])
            return merge(left_arr, right_arr)
        nums = divide(nums)
        return self.count
```

### 5.Quick Sort

```python
def quick_sort(arr):
    if len(arr)<=1:
        return arr
    else:
        pivot = arr[len(arr)//2] #middle element as pivot

        left = [x for x in arr if x<pivot] #elements less than pivot
        middle = [x for x in arr if x==pivot] #elements equal to pivot
        right = [x for x in arr if x>pivot] #elements greater than pivot

        return quick_sort(left) + middle + quick_sort(right) #recursively sort the left and right parts of the array
    
arr = [10,7,8,9,1,5]
sorted_arr = quick_sort(arr)
print("Sorted array is:", sorted_arr)
```

```python
#Partition Algorithm
def partition(arr,low,high):
    pivot = arr[high]
    i = low-1 #index of smaller element
    for j in range(low, high):
        if arr[j]<=pivot:
            i+=1
            arr[i], arr[j] = arr[j], arr[i] #swap
    arr[i+1], arr[high] = arr[high], arr[i+1] #swap
    return i+1
def quick_sort_recursive(arr,low,high):
    if low<high:
        pi = partition(arr,low,high)
        quick_sort_recursive(arr,low,pi-1)
        quick_sort_recursive(arr,pi+1,high)

arr = [10,7,8,9,1,5]
quick_sort_recursive(arr,0,len(arr)-1)
print("Sorted array is:", arr)
```

```python
'''3. Given an array where all its elements are sorted in increasing order except two swapped
 elements, sort it in linear time. Assume there are no duplicates in the array.
 Input: A[] = [3, 8, 6, 7, 5, 9, 10]
 Output: A[] = [3, 5, 6, 7, 8, 9, 10]'''

def fix_swapped_elements(arr):
    n = len(arr)
    first = second = -1
    # Identify the two swapped elements
    for i in range(n - 1):
        if arr[i] > arr[i + 1]:
            if first == -1:
                first = i
            second = i + 1
    # Swap the two identified elements
    arr[first], arr[second] = arr[second], arr[first]
# Test the function
arr = [3, 8, 6, 7, 5, 9, 10]
print("Original array:", arr)
fix_swapped_elements(arr)
print("Sorted array:", arr)
```

### 6.Cycle Sort

```python
def cycle_sort(array):
    for cycle_sort in range(0,len(array)-1):
        item = array[cycle_sort]
        pos = cycle_sort
        for i in range(cycle_sort+1,len(array)):
            if array[i]<item:
                pos+=1
        if pos == cycle_sort:
            continue
        while item == array[pos]:
            pos+=1
        array[pos],item = item,array[pos]
        while pos != cycle_sort:
            pos = cycle_sort
            for i in range(cycle_sort+1,len(array)):
                if array[i]<item:
                    pos+=1
            while item == array[pos]:
                pos+=1
            array[pos],item = item,array[pos]

arr = [4,3,6,3,3,3,6,1]
cycle_sort(arr)
print(arr)
```

```python
class Node:
    def __init__(self, value):
        self.data = value
        self.next = None
```

### Creating a Linked List

```python
class LinkedList:
    def __init__(self):
        self.head = None

    # Adding node to LinkedList
    def add_node(self, value):
        new_node = Node(value)

        # Creating head node
        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next is not None:
                current = current.next
            current.next = new_node

l1 = LinkedList()

l1.add_node(value= 10)
l1.add_node(value=15)
l1.add_node(value=20)
```

# Doubly Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data # value to be stored in the node
        self.next = None
        self.prev = None  # pointer to the previous node


# Doubly Linked List
class DoublyLinkedList:
    def __init__(self):
        self.head = None # pointer to the first node

    # Insert at the start
    def insert_at_start(self, data):
        new_node = Node(data) # create a new node
        if self.head is None:
            self.head = new_node # if the list is empty, set the new node as the head
        else:
            new_node.next = self.head # set the next pointer of the new node to the current head
            self.head.prev = new_node # set the previous pointer of the current head to the new node
            self.head = new_node  # set the new node as the head

    # Insert at the end
    def insert_at_end(self, data):
        new_node = Node(data)
        if self.head is None: # if the list is empty, set the new node as the head
            self.head = new_node
        else:
            temp = self.head  # traverse the list to find the last node
            while temp.next is not None:  # until the last node is reached
                temp = temp.next  # move to the next node
            temp.next = new_node
            new_node.prev = temp  # set the previous pointer of the new node to the last node

    # Insert at a specific position
    def insert_at_position(self, data, position):
        new_node = Node(data)
        if position == 0: # if the position is 0, insert at the start
            self.insert_at_start(data)  # call the insert_at_start method
        else:
            temp = self.head
            for _ in range(position - 1): # traverse the list to find the node at the given position
                temp = temp.next
                if temp is None:
                    raise IndexError("Position out of bounds") # raise an error if the position is out of bounds
            new_node.next = temp.next  # set the next pointer of the new node to the node at the given position
            new_node.prev = temp # set the previous pointer of the new node to the node at the given position
            if temp.next:
                temp.next.prev = new_node # set the previous pointer of the next node to the new node
            temp.next = new_node

    # Delete from the start
    def delete_from_start(self):
        if self.head is None: 
            return
        if self.head.next is None: # if there is only one node in the list
            self.head = None
        else:
            self.head = self.head.next # set the next node as the head
            self.head.prev = None

    # Delete from the end
    def delete_from_end(self):
        if self.head is None: # if the list is empty, return
            return
        if self.head.next is None:
            self.head = None # if there is only one node in the list, set the head to None
        else:
            temp = self.head # traverse the list to find the last node
            while temp.next is not None: # until the last node is reached
                temp = temp.next # move to the next node
            temp.prev.next = None  # set the next pointer of the second last node to None

    # Delete a node by value
    def delete_by_value(self, data):
        if self.head is None: # if the list is empty,
            return
        temp = self.head # traverse the list to find the node with the given value
        while temp is not None: # until the end of the list is reached
            if temp.data == data:
                if temp.prev: # if the node to be deleted is not the first node
                    temp.prev.next = temp.next # set the next pointer of the previous node to the next node
                if temp.next:
                    temp.next.prev = temp.prev # set the previous pointer of the next node to the previous node
                if temp == self.head:  # if head is to be deleted
                    self.head = temp.next # set the next node as the head
                break
            temp = temp.next # move to the next node

    # Utility function to print the list
    def print_list(self):
        temp = self.head # start from the head
        while temp:
            print(temp.data, end=" <-> ") # print the data of the current node
            temp = temp.next
        print("None")
class Node:
    def __init__(self, data):  # Node class
        self.data = data
        self.next = None
        self.prev = None

        

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    # Insert at the start
    def insert_at_start(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node

    # Insert at the end
    def insert_at_end(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            temp = self.head
            while temp.next is not None:
                temp = temp.next
            temp.next = new_node
            new_node.prev = temp

    # Insert at a specific position
    def insert_at_position(self, data, position):
        new_node = Node(data)
        if position == 0:
            self.insert_at_start(data)
        else:
            temp = self.head
            for _ in range(position - 1):
                temp = temp.next
                if temp is None:
                    raise IndexError("Position out of bounds")
            new_node.next = temp.next
            new_node.prev = temp
            if temp.next:
                temp.next.prev = new_node
            temp.next = new_node

    # Delete from the start
    def delete_from_start(self):
        if self.head is None:
            return
        if self.head.next is None:
            self.head = None
        else:
            self.head = self.head.next
            self.head.prev = None

    # Delete from the end
    def delete_from_end(self):
        if self.head is None:
            return
        if self.head.next is None:
            self.head = None
        else:
            temp = self.head
            while temp.next is not None:
                temp = temp.next
            temp.prev.next = None

    # Delete a node by value
    def delete_by_value(self, data):
        if self.head is None:
            return
        temp = self.head
        while temp is not None:
            if temp.data == data:
                if temp.prev:
                    temp.prev.next = temp.next
                if temp.next:
                    temp.next.prev = temp.prev
                if temp == self.head:  # if head is to be deleted
                    self.head = temp.next
                break
            temp = temp.next

    # Utility function to print the list
    def print_list(self):
        temp = self.head
        while temp:
            print(temp.data, end=" <-> ")
            temp = temp.next
        print("None")


a1 = DoublyLinkedList()
a1.insert_at_start(1)
a1.insert_at_end(2)
a1.insert_at_end(3)
a1.insert_at_end(4)
a1.insert_at_end(5)
a1.print_list()

```

# Find the middle linked list

```python
from typing import Optional
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next



class Solution:
    def middleNode(self, head: Optional[ListNode]) -> Optional[ListNode]:

        # Edge case
        if head ==None or head.next == None:
            return head
        
        length = 1
        current = head

        # finding length of the array
        while current.next != None:
            current = current.next
            length+=1
        middle = length//2

        # updating head to middle of the Linked List
        while middle:
            head = head.next
            middle -= 1
        return head
    
# Helper functions to create and display
def create_linked_list(arr):
    head = ListNode(arr[0])
    current = head
    for val in arr[1:]:
        current.next = ListNode(val)
        current = current.next
    return head
def display_linked_list(head):
    current = head
    while current:
        print(current.val, end=" -> ")
        current = current.next
    print("None")
 # Example usage
head = create_linked_list([1, 2, 3, 4, 5])
sol = Solution()
new_head = sol.middleNode(head)
display_linked_list(new_head)   

```

```python
# Without using any space


class Solution:
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> Optional[ListNode]:

        # find the len of A and B
        lenA, lenB = 1,1
        currA, currB = headA, headB

        while currA.next:
            lenA+=1
            currA = currA.next

        while currB.next:
            lenB+=1
            currB = currB.next

        diff = abs(lenA - lenB)
        
        # skipping nodes
        if lenA>lenB:
            while diff:
                headA= headA.next
                diff -= 1
        elif lenB>lenA:
            while diff:
                headB = headB.next
                diff -=1
        while(headA != headB):
            headA = headA.next
            headB = headB.next

        return headA
    
# Helper functions to create and display
def create_linked_list(arr):
    head = ListNode(arr[0])
    current = head
    for val in arr[1:]:
        current.next = ListNode(val)
        current = current.next
    return head

def display_linked_list(head):
    current = head
    while current:
        print(current.val, end=" -> ")
        current = current.next
    print("None")
 # Example usage
head = create_linked_list([1, 2, 3, 4, 5])
sol = Solution()
new_head = sol.getIntersectionNode(head, head)
display_linked_list(new_head)
```

# Merge two sorted linked list

```python
class Solution:
    def mergeTwoLists( list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        dummy = ListNode()
        tail = dummy
        while list1 and list2:
            if list1.val < list2.val:
                tail.next = list1
                list1 = list1.next
            else:
                tail.next = list2
                list2 = list2.next
            tail = tail.next

        if list1 is not None:
            tail.next = list1
        else:
            tail.next = list2

        return dummy.next
    
# Helper functions to create and display
def create_linked_list(arr):
    head = ListNode(arr[0])
    current = head
    for val in arr[1:]:
        current.next = ListNode(val)
        current = current.next
    return head

def display_linked_list(head):
    current = head
    while current:
        print(current.val, end=" -> ")
        current = current.next
    print("None")
 # Example usage
head = create_linked_list([1, 2, 3, 4, 5])
sol = Solution()
new_head = sol.mergeTwoLists(head, head)
display_linked_list(new_head)
```

```python
def get_middle(head):
    if not head:
        return head
    slow = head
    fast = head

    while fast.next and fast.next.next:
        slow = slow.next
        fast = fast.next.next
    return slow

def merge_sort(head):
    if not head or not head.next:
        return head
    
    mid = get_middle(head)
    # break the linked list
    next_to_mid = mid.next
    mid.next = None

    left = merge_sort(head)
    right = merge_sort(next_to_mid)

    # merge two sorted linked lists
    sorted_list = mergeTwoLists(left,right)
    return sorted_list

# Helper functions to create and display
def create_linked_list(arr):
    head = ListNode(arr[0])
    current = head
    for val in arr[1:]:
        current.next = ListNode(val)
        current = current.next
    return head

def display_linked_list(head):
    current = head
    while current:
        print(current.val, end=" -> ")
        current = current.next
    print("None")
 # Example usage
head = create_linked_list([1, 2, 3, 4, 5])
sol = Solution()
new_head = sol.merge_sort(head)
display_linked_list(new_head)
```

# Stack

```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None


class Stack:
    def __init__(self) -> None:
        self.head = None

    def push(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def pop(self):
        if self.head is None:
            raise IndexError("stack is empty")
        popped_value = self.head.data
        self.head = self.head.next
        return popped_value

    def print(self):
        while self.head is not None:
            print(self.head.data)
            self.head = self.head.next


if __name__ == "__main__":
    s = Stack()
    s.push(4)
    s.push(8)
    s.push(9)
    s.pop()
    s.pop()
    s.print()
```

```python
def precendence(operator):
    if operator == "+" or operator == "-":
        return 1
    elif operator == "*" or operator == "/":
        return 2
    elif operator == "^":
        return 3
    return 0


def infixToPost(expression):
    stack = []  # stack to hold the operators
    output = []  # output to store the output expr

    for char in expression:
        if char.isalnum():  # [0,9 or a-z or A-Z]
            output.append(char)
        elif char == "(":
            stack.append(char)
        elif char == ")":
            while stack and stack[-1] != "(":
                output.append(stack.pop())
            stack.pop()  # pop '(' from the stack
        else: 
            while stack and precendence(stack[-1])>=precendence(char):
                output.append(stack.pop())
            stack.append(char)     
    while stack:
            output.append(stack.pop())
    return "".join(output)    

infix="(A+B)/C" 
postfix=infixToPost(infix)
print(postfix)   
            
```

# Queue

```python
# Node class represents each element in the queue
class Node:
    def __init__(self, data):
        self.data = data  
        self.next = None  


class Queue:
    def __init__(self):
        self.front = None  
        self.rear = None   

    def is_empty(self):
        return self.front is None

    def enqueue(self, data):
        new_node = Node(data)  # Create a new node
        if self.rear is None:
            self.front = self.rear = new_node
            return
        self.rear.next = new_node  # Link the new node with the last node
        self.rear = new_node  # Move the rear to the new node

    def dequeue(self):
        if self.is_empty():
            return None
        temp = self.front  # Store the front node
        self.front = temp.next  # Move front to the next node
        if self.front is None:
            self.rear = None  # If queue becomes empty, set rear to None
        return temp.data

    def peek(self): # Get the front element of the queue
        if self.is_empty(): # If the queue is empty, return None
            return None #
        return self.front.data # Return the front element

    def print_queue(self):
        temp = self.front # Start from the front
        while temp:
            print(temp.data, end=" ")
            temp = temp.next
        print()


# Example usage:
if __name__ == '__main__': # Main function
    queue = Queue()

    # Enqueue some elements
    queue.enqueue(1)
    queue.enqueue(2)
    queue.enqueue(3)
    queue.enqueue(4)

    print("Queue:")
    queue.print_queue()

    # Dequeue elements and print
    print("Dequeue:", queue.dequeue())
    print("Dequeue:", queue.dequeue())

    print("Queue after dequeue:")
    queue.print_queue()

    # Peek at the front element
    print("Front of the queue:", queue.peek())
```

```python
from collections import deque

def reverse_queue(q):
    stack=[]
    
    #push elements from the queue to stack
    while q:
        stack.append(q.popleft())
    
    #pop elements from the stack to the queue
    
    while stack:
        q.append(stack.pop())
    return q    

queue= reverse_queue(deque([1,2,3,4,5]))
print(queue)            
```

```python
#Creating node class
class Node:
    def __init__(self, value):
        self.val = value
        self.left = None
        self.right = None
        
```

```python
class BinaryTreeOperations:
    @staticmethod
    def display(root):
        if root is None:
            return
        print(root.val, end=' ')
        BinaryTreeOperations.display(root.left)
        BinaryTreeOperations.display(root.right)
    @staticmethod
    def sum_of_nodes(root):
        if root is None:
            return 0
        return root.val + BinaryTreeOperations.sum_of_nodes(root.left) + BinaryTreeOperations.sum_of_nodes(root.right)
    @staticmethod
    def find_max_node(root):
        if root is None or (root.left is None and root.right is None):
            return root
        max_left = BinaryTreeOperations.find_max_node(root.left)
        max_right = BinaryTreeOperations.find_max_node(root.right)
        return max(max_left, max_right, root, key=lambda x:x.val if x else float('-inf'))
    @staticmethod
    def height(root):
        if root is None:
            return -1
        return 1 + max(BinaryTreeOperations.height(root.left),BinaryTreeOperations.height(root.right))
    @staticmethod
    def size_of_tree(root):
        if root is None:
            return 0
        return 1 + BinaryTreeOperations.size_of_tree(root.left) +BinaryTreeOperations.size_of_tree(root.right)
```

```python
# Preorder traversal

def preorder(root):
    if root is None:
        return
    print(root.val, end = " ")
    preorder(root.left)
    preorder(root.right)

preorder(root)
```

```python
# Inorder traversal

def inorder(root):
    if root is None:
        return
    
    inorder(root.left)
    print(root.val, end = " ")
    inorder(root.right)

inorder(root)
```

```python
# Postorder traversal

def postorder(root):
    if root is None:
        return
    
    postorder(root.left)
    postorder(root.right)
    print(root.val, end = " ")

postorder(root)
```

```python
from collections import deque

def level_order(root):
    if root is None:
        return
    queue = deque([root])

    while queue:
        node = queue.popleft()
        print(node.val, end= " ")
        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)

level_order(root)
```

```python
#Print nth level
from collections import deque

def print_nth_level(root, n):
    if root is None:
        return
    
    level = 0
    queue = deque([root])

    while queue:
        level_size = len(queue)
        if level == n:
            for i in range(level_size):
                node = queue.popleft()
                print(node.val, end = " ")
            return
        
        for _ in range(level_size):
            node = queue.popleft()
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)

        level+=1

    print(" Level not found ")

        
print_nth_level(root, 2)
print()
print_nth_level(root, n=1)
```
