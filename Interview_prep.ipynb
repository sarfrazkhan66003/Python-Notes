```python
#Write a recursive function to find factorial of a number.
def factorial(n):
    if n == 0 or n == 1:  
        return 1
    return n * factorial(n - 1)

print(factorial(5))   # Output: 120
# Time Complexity: O(n)
# Space Complexity: O(n) due to recursion stack
```

```python
#Return the Nth Fibonacci number
def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)

print(fib(6))  # Output: 8
# Time Complexity: O(2^n)
# Space Complexity: O(n) due to recursion stack
```

```python
#Sum of digits using recursion
def sum_digits(n):
    if n == 0:
        return 0
    return (n % 10) + sum_digits(n // 10)

print(sum_digits(12345))   # Output: 15
# Time Complexity: O(d) where d is the number of digits
# Space Complexity: O(d) due to recursion stack
```

```python
#Reverse a String Using Recursion
def reverse_string(s):
    if len(s) == 0:
        return s
    return reverse_string(s[1:]) + s[0]

print(reverse_string("Adnan"))   # "zarfras"
# Time Complexity: O(n)
# Space Complexity: O(n) due to recursion stack
```

```python
#Check Palindrome Using Recursion
def is_palindrome(s):
    if len(s) <= 1:
        return True
    if s[0] != s[-1]:
        return False
    return is_palindrome(s[1:-1])

print(is_palindrome("madam"))   # True
# Time Complexity: O(n)
# Space Complexity: O(n) due to recursion stack
```

```python
#Power of a Number (xâ¿) using Recursion
def power(x, n):
    if n == 0:
        return 1
    return x * power(x, n-1)

print(power(2, 5))  # 32
# Time Complexity: O(n)
# Space Complexity: O(n) due to recursion stack
```

```python
#Find GCD (Euclid Algorithm Recursion)
def gcd(a, b):
    if b == 0:
        return a
    return gcd(b, a % b)

print(gcd(48, 18))  # 6
# Time Complexity: O(log(min(a, b)))
# Space Complexity: O(log(min(a, b))) due to recursion stack
```

```python
#Print 1 to N Using Recursion
def print_1_to_n(n):
    if n == 0:
        return
    print_1_to_n(n-1)
    print(n)

print_1_to_n(5)
# Time Complexity: O(n)
# Space Complexity: O(n) due to recursion stack
```

```python
#Print N to 1 (Reverse Order)
def print_n_to_1(n):
    if n == 0:
        return
    print(n)
    print_n_to_1(n-1)

print_n_to_1(5)
# Time Complexity: O(n)
# Space Complexity: O(n) due to recursion stack
```

```python
#Count Number of Zeros in a Number (Recursion)
def count_zero(n):
    if n == 0:
        return 1   # special case: number is 0
    def helper(x):
        if x == 0:
            return 0
        return (1 if x % 10 == 0 else 0) + helper(x // 10)
    return helper(n)

print(count_zero(10203040))   # 4
# Time Complexity: O(d) where d is the number of digits
# Space Complexity: O(d) due to recursion stack
```

```python
#Binary search (recursive)
def bsearch(arr, l, r, x):
    if l > r:
        return -1
    mid = (l + r)//2
    if arr[mid] == x:
        return mid
    if arr[mid] > x:
        return bsearch(arr, l, mid-1, x)
    return bsearch(arr, mid+1, r, x)

bsearch([2, 3, 4, 10, 40], 0, 4, 10)
# A simple implementation of binary search in Python
# Returns index of x in arr if present, else -1
# Time Complexity: O(log n)
# Space Complexity: O(log n) due to recursion stack
```

```python
#First occurrence in array
def first_occ(a, i, x):
    if i == len(a):
        return -1
    if a[i] == x:
        return i
    return first_occ(a, i+1, x)
print(first_occ([1, 2, 3, 2, 4], 0, 2))  # Output: 1
# Time Complexity: O(n)
# Space Complexity: O(n) due to recursion stack
```

```python
#Last occurrence in array
def last_occ(a, i, x):
    if i == len(a):
        return -1
    rest = last_occ(a, i+1, x)
    if rest != -1:
        return rest
    if a[i] == x:
        return i
    return -1
print(last_occ([1, 2, 3, 2, 4], 0, 2))  # Output: 3
# Time Complexity: O(n) 
# Space Complexity: O(n) due to recursion stack
```

```python
#Check sorted array
def is_sorted(a, i=0):
    if i == len(a)-1:
        return True
    return a[i] <= a[i+1] and is_sorted(a, i+1)
print(is_sorted([1, 2, 3, 4, 5]))  # True
print(is_sorted([1, 3, 2, 4, 5]))  # False
# Time Complexity: O(n) 
# Space Complexity: O(n) due to recursion stack
```

```python
#Count subsets with sum = k
def count_subsets(a, n, k):
    if k == 0:
        return 1
    if n == 0:
        return 0
    return count_subsets(a, n-1, k) + count_subsets(a, n-1, k-a[n-1])
print(count_subsets([2, 3, 5, 6, 8, 10], 6, 10))  # Output: 3
# Time Complexity: O(2^n)
# Space Complexity: O(n) due to recursion stack
```

```python
#Tower of Hanoi
def hanoi(n, src, aux, dst):
    if n == 1:
        print(f"{src} -> {dst}")
        return
    hanoi(n-1, src, dst, aux)
    print(f"{src} -> {dst}")
    hanoi(n-1, aux, src, dst)

hanoi(3, 'A', 'B', 'C')
# Time Complexity: O(2^n)
# Space Complexity: O(n) due to recursion stack
```

```python
#Move N disks from source â†’ destination using helper rod.
def tower_of_hanoi(n, source, helper, target):
    if n == 1:
        print(f"Move disk 1 from {source} to {target}")
        return
    
    tower_of_hanoi(n-1, source, target, helper)
    print(f"Move disk {n} from {source} to {target}")
    tower_of_hanoi(n-1, helper, source, target)

tower_of_hanoi(3, 'A', 'B', 'C')
# Time Complexity: O(2^n)
# Space Complexity: O(n) due to recursion stack
```

```python
#Generate All Permutations of a String
def permute(s, ans=""):
    if len(s) == 0:
        print(ans)
        return
    for i in range(len(s)):
        permute(s[:i] + s[i+1:], ans + s[i])

permute("abc")
# Time Complexity: O(n!)
# Space Complexity: O(n) due to recursion stack
```

```python
#Balanced Parentheses
def gen_paren(o, c, s=""):
    if o == 0 and c == 0:
        print(s)
        return
    if o > 0:
        gen_paren(o-1, c, s+"(")
    if c > o:
        gen_paren(o, c-1, s+")")
gen_paren(3, 3)
# Time Complexity: O(4^n / sqrt(n)) (Catalan number)
# Space Complexity: O(n) due to recursion stack
```

```python
#Generate Balanced Parentheses (N-Pairs)
def generate_parenthesis(open, close, s=""):
    if open == 0 and close == 0:
        print(s)
        return
    
    if open > 0:
        generate_parenthesis(open-1, close, s+"(")
    if close > open:
        generate_parenthesis(open, close-1, s+")")

generate_parenthesis(3, 3)
# Time Complexity: O(4^n / sqrt(n)) (Catalan number)
# Space Complexity: O(n) due to recursion stack
```

```python
#N-Queens
def nqueens(n):
    board = [-1]*n
    def safe(r, c):
        for i in range(r):
            if board[i] == c or abs(board[i]-c)==abs(i-r):
                return False
        return True
    def solve(r):
        if r == n:
            print(board)
            return
        for c in range(n):
            if safe(r, c):
                board[r] = c
                solve(r+1)
    solve(0)
nqueens(4)
# Time Complexity: O(N!)
# Space Complexity: O(N) due to recursion stack
```

```python
def is_safe(board, row, col, n):
    for i in range(row):
        if board[i] == col or abs(board[i] - col) == abs(i - row):
            return False
    return True

def solve_n_queens(row, board, n):
    if row == n:
        print(board)
        return
    
    for col in range(n):
        if is_safe(board, row, col, n):
            board[row] = col
            solve_n_queens(row+1, board, n)

solve_n_queens(0, [-1]*4, 4)   # For 4x4
# Time Complexity: O(N!)
# Space Complexity: O(N) due to recursion stack
```

```python
#Unique Permutations
def perm_unique(nums):
    nums.sort()
    used=[False]*len(nums)
    def solve(path):
        if len(path)==len(nums):
            print(path)
            return
        for i in range(len(nums)):
            if used[i]: continue
            if i>0 and nums[i]==nums[i-1] and not used[i-1]:
                continue
            used[i]=True
            solve(path+[nums[i]])
            used[i]=False
    solve([])
perm_unique([1, 1, 2])
# Time Complexity: O(n!)    
# Space Complexity: O(n) due to recursion stack
```

```python
#Subset Sum
def subset_sum(a, i, curr, target):
    if curr == target:
        print("Found subset")
        return True
    if i == len(a):
        return False
    return subset_sum(a, i+1, curr+a[i], target) or \
           subset_sum(a, i+1, curr, target)
print(subset_sum([3, 34, 4, 12, 5, 2], 0, 0, 9))  # True
# Time Complexity: O(2^n)   
# Space Complexity: O(n) due to recursion stack
```

```python
#K-th Symbol in Grammar
def kth(n, k):
    if n == 1:
        return 0
    mid = 2**(n-1)//2
    if k <= mid:
        return kth(n-1, k)
    return 1 - kth(n-1, k-mid)
print(kth(4, 5))  # Output: 1
# Time Complexity: O(n)
# Space Complexity: O(n) due to recursion stack
```

```python
#Gray Code
def gray(n):
    if n == 0:
        return ["0"]
    if n == 1:
        return ["0","1"]
    prev = gray(n-1)
    res = ["0"+p for p in prev] + ["1"+p for p in reversed(prev)]
    return res
print(gray(3))  # Output: ['000', '001', '011', '010', '110', '111', '101', '100']
# Time Complexity: O(2^n)
# Space Complexity: O(2^n)
```

```python
#Decode Ways
def decode(s):
    if len(s) == 0:
        return 1
    if s[0] == "0":
        return 0
    res = decode(s[1:])
    if len(s)>=2 and int(s[:2])<=26:
        res += decode(s[2:])
    return res
print(decode("226"))  # Output: 3
# Time Complexity: O(2^n)   
# Space Complexity: O(n) due to recursion stack
```

```python
#Generate IP Addresses
def ip(s, parts=0, curr=""):
    if parts==4 and s=="":
        print(curr[:-1]); return
    if parts==4:
        return
    for i in range(1,4):
        if i <= len(s):
            seg=s[:i]
            if seg[0]=='0' and len(seg)>1: continue
            if int(seg)<=255:
                ip(s[i:], parts+1, curr+seg+".")
ip("25525511135")
# Time Complexity: O(1) since max 12 digits 
# Space Complexity: O(1) due to recursion stack
```

```python
#Restore IP Addresses
def restore_ip(s, parts=0, curr="", result=[]):
    if parts == 4 and not s:
        print(curr[:-1])
        return
    if parts == 4 or not s:
        return

    for i in range(1, 4):
        if i <= len(s) and int(s[:i]) <= 255 and (i == 1 or s[0] != "0"):
            restore_ip(s[i:], parts+1, curr + s[:i] + ".", result)
restore_ip("25525511135")
# Time Complexity: O(1) since max 12 digits
# Space Complexity: O(1) due to recursion stack
```

```python
#Word Search in a Grid (DFS + Recursion)
def dfs(board, word, i, j, index):
    if index == len(word):
        return True
    if i<0 or j<0 or i>=len(board) or j>=len(board[0]) or board[i][j] != word[index]:
        return False

    temp = board[i][j]
    board[i][j] = "#"

    found = (dfs(board, word, i+1, j, index+1) or
             dfs(board, word, i-1, j, index+1) or
             dfs(board, word, i, j+1, index+1) or
             dfs(board, word, i, j-1, index+1))

    board[i][j] = temp
    return found

board = [
    ['A','B','C','E'],
    ['S','F','C','S'],
    ['A','D','E','E']
]
word = "ABCCED"
print(dfs(board, word, 0, 0, 0))  # Output: True
# Time Complexity: O(M * N * 3^L) where MxN is board size and L is word length
# Space Complexity: O(L) due to recursion stack
```

```python
#Count All Paths in a Grid (Recursion + DP)
def count_paths(m, n):
    if m == 1 or n == 1:
        return 1
    return count_paths(m-1, n) + count_paths(m, n-1)

print(count_paths(3, 3))
# Time Complexity: O(2^(m+n))
# Space Complexity: O(m+n) due to recursion stack
```

```python
#Letter Combinations of Phone Number
mapping = {
    "2": "abc", "3": "def", "4": "ghi",
    "5": "jkl", "6": "mno", "7": "pqrs",
    "8": "tuv", "9": "wxyz"
}

def phone_combinations(digits, s=""):
    if len(digits) == 0:
        print(s)
        return

    letters = mapping[digits[0]]
    for ch in letters:
        phone_combinations(digits[1:], s + ch)

phone_combinations("23")
# Time Complexity: O(3^n * 4^m) where n is number of digits mapping to 3 letters and m to 4 letters
# Space Complexity: O(n) due to recursion stack
```

```python
#K-th Symbol in Grammar (Google Favorite)
def kth_symbol(n, k):
    if n == 1:
        return 0
    mid = 2**(n-1) // 2
    
    if k <= mid:
        return kth_symbol(n-1, k)
    return 1 - kth_symbol(n-1, k-mid)

print(kth_symbol(4, 5))
# Output: 1
# Time Complexity: O(n)
# Space Complexity: O(n) due to recursion stack
```

```python
#Josephus Problem (Recursion)
def josephus(n, k):
    if n == 1:
        return 0
    return (josephus(n-1, k) + k) % n

print(josephus(7, 3))
# Output: 3
# Time Complexity: O(n)
# Space Complexity: O(n) due to recursion stack
```

```python
#Print All Palindromic Partitions of a String
def is_pal(s):
    return s == s[::-1]

def pal_partitions(s, ans=[]):
    if not s:
        print(ans)
        return

    for i in range(1, len(s)+1):
        part = s[:i]
        if is_pal(part):
            pal_partitions(s[i:], ans + [part])
pal_partitions("aab")
# Output: ['a', 'a', 'b'], ['aa', 'b']
# Time Complexity: O(n * 2^n)
# Space Complexity: O(n) due to recursion stack
```

```python
#Word Search in Grid
def exist(board, word):
    R,C=len(board),len(board[0])
    def dfs(i,j,k):
        if k==len(word): return True
        if i<0 or j<0 or i>=R or j>=C or board[i][j]!=word[k]:
            return False
        temp=board[i][j]
        board[i][j]="#"
        found = dfs(i+1,j,k+1) or dfs(i-1,j,k+1) or dfs(i,j+1,k+1) or dfs(i,j-1,k+1)
        board[i][j]=temp
        return found
    for i in range(R):
        for j in range(C):
            if dfs(i,j,0): return True
    return False
print(exist([['A','B','C','E'],
             ['S','F','C','S'],
             ['A','D','E','E']], "ABCCED"))  # Output: True
# Time Complexity: O(M * N * 3^L) where MxN is board size and L is word length
# Space Complexity: O(L) due to recursion stack
```

```python
#. Word Break (all solutions)
def wb(s, dic):
    res=[]
    def dfs(i, cur):
        if i==len(s):
            res.append(" ".join(cur)); return
        for j in range(i+1, len(s)+1):
            if s[i:j] in dic:
                dfs(j, cur+[s[i:j]])
    dfs(0, [])
    return res
print(wb("catsanddog", {"cat", "cats", "and", "sand", "dog"}))  # Output: ["cats and dog", "cat sand dog"]
# Time Complexity: O(2^n) in worst case
# Space Complexity: O(n) due to recursion stack
```

```python
#Palindromic Partitioning
def pal_part(s):
    res=[]
    def ispal(x): return x==x[::-1]
    def dfs(i,path):
        if i==len(s):
            res.append(path.copy()); return
        for j in range(i+1,len(s)+1):
            if ispal(s[i:j]):
                dfs(j,path+[s[i:j]])
    dfs(0,[])
    return res
print(pal_part("aab"))  # Output: [['a', 'a', 'b'], ['aa', 'b']]
# Time Complexity: O(n * 2^n)
```

```python
#Solve Sudoku
def sudoku(board):
    def valid(r,c,v):
        for k in range(9):
            if board[r][k]==v or board[k][c]==v:
                return False
        sr,sc=3*(r//3),3*(c//3)
        for i in range(sr,sr+3):
            for j in range(sc,sc+3):
                if board[i][j]==v:
                    return False
        return True
    def solve():
        for i in range(9):
            for j in range(9):
                if board[i][j]==0:
                    for v in range(1,10):
                        if valid(i,j,v):
                            board[i][j]=v
                            if solve(): return True
                            board[i][j]=0
                    return False
        return True
    solve()
    return board 
solved = sudoku([[5,3,0,0,7,0,0,0,0],
        [6,0,0,1,9,5,0,0,0],
        [0,9,8,0,0,0,0,6,0],
        [8,0,0,0,6,0,0,0,3],
        [4,0,0,8,0,3,0,0,1],
        [7,0,0,0,2,0,0,0,6],
        [0,6,0,0,0,0,2,8,0],
        [0,0,0,4,1,9,0,0,5],
        [0,0,0,0,8,0,0,7,9]])
for row in solved:
    print(row)

# Time Complexity: O(9^(n*n)) in worst case
# Space Complexity: O(n*n) due to recursion stack
```

```python
#Unique Paths with Obstacles
def unique_paths_with_obstacles(obstacle_grid):
    m, n = len(obstacle_grid), len(obstacle_grid[0])
    
    def paths(i, j):
        if i >= m or j >= n or obstacle_grid[i][j] == 1:
            return 0
        if i == m - 1 and j == n - 1:
            return 1
        return paths(i + 1, j) + paths(i, j + 1)
    
    return paths(0, 0)
print(unique_paths_with_obstacles([[0,0,0],[0,1,0],[0,0,0]]))  # Output: 2
# Time Complexity: O(2^(m+n))
# Space Complexity: O(m+n) due to recursion stack
```

```python
#N-Queens â€” generate all solutions
def n_queens(n):
    res=[]
    cols=set(); diag=set(); anti=set()
    board=[-1]*n
    def place(r):
        if r==n:
            res.append(board.copy())
            return
        for c in range(n):
            if c in cols or (r-c) in diag or (r+c) in anti: continue
            cols.add(c); diag.add(r-c); anti.add(r+c); board[r]=c
            place(r+1)
            cols.remove(c); diag.remove(r-c); anti.remove(r+c)
    place(0)
    # convert to board strings if desired
    return res
print(n_queens(4))
# Time Complexity: O(N!)
# Space Complexity: O(N) due to recursion stack
```

```python
#Word Break II â€” all sentences (with memo)
def word_break_all(s, wordDict):
    wordset=set(wordDict)
    memo={}
    def dfs(i):
        if i==len(s): return [""]
        if i in memo: return memo[i]
        ans=[]
        for j in range(i+1, len(s)+1):
            w=s[i:j]
            if w in wordset:
                for rest in dfs(j):
                    ans.append(w + ("" if rest=="" else " "+rest))
        memo[i]=ans
        return ans
    return dfs(0)
print(word_break_all("catsanddog", ["cat","cats","and","sand","dog"]))
# Time Complexity: O(n^3) in worst case
# Space Complexity: O(n^2) due to memoization and recursion stack
```

```python
#All palindromic partitions of a string
def pal_partitions(s):
    res=[]
    def is_pal(x): return x==x[::-1]
    def dfs(i, path):
        if i==len(s):
            res.append(path.copy()); return
        for j in range(i+1, len(s)+1):
            part=s[i:j]
            if is_pal(part):
                path.append(part)
                dfs(j, path)
                path.pop()
    dfs(0, [])
    return res
print(pal_partitions("aab"))
# Time Complexity: O(n * 2^n)   
# Space Complexity: O(n) due to recursion stack
```

```python
#K-equal sum subsets (partition to k subsets)
def can_partition_k_subsets(nums, k):
    total=sum(nums)
    if total % k != 0: return False
    target=total // k
    nums.sort(reverse=True)
    buckets=[0]*k
    def dfs(i):
        if i==len(nums): return True
        for j in range(k):
            if buckets[j] + nums[i] <= target:
                buckets[j] += nums[i]
                if dfs(i+1): return True
                buckets[j] -= nums[i]
            if buckets[j]==0: break
        return False
    return dfs(0)
print(can_partition_k_subsets([4,3,2,3,5,2,1], 4))
# Time Complexity: O(k^n) in worst case
# Space Complexity: O(n) due to recursion stack
```

```python
def can_partition_k_subsets(nums, k):
    total=sum(nums)
    if total%k!=0: return False
    target=total//k
    nums.sort(reverse=True)
    if nums[0]>target: return False
    buckets=[0]*k
    def dfs(i):
        if i==len(nums):
            return all(b==target for b in buckets)
        v=nums[i]
        seen=set()
        for j in range(k):
            if buckets[j]+v<=target and buckets[j] not in seen:
                buckets[j]+=v
                if dfs(i+1): return True
                buckets[j]-=v
                seen.add(buckets[j])
            if buckets[j]==0: break
        return False
    return dfs(0)
print(can_partition_k_subsets([4,3,2,3,5,2,1], 4))
# Time Complexity: O(k^n) in worst case
# Space Complexity: O(n) due to recursion stack
```

```python
#All Hamiltonian paths in an undirected graph
def hamiltonian_paths(adj, start=None):
    n=len(adj)
    res=[]
    if start is None:
        starts=range(n)
    else:
        starts=[start]
    def dfs(u, visited, path):
        if len(path)==n:
            res.append(path.copy()); return
        for v in adj[u]:
            if v not in visited:
                visited.add(v); path.append(v)
                dfs(v, visited, path)
                path.pop(); visited.remove(v)
    for s in starts:
        dfs(s, {s}, [s])
    return res

# adj is list of neighbor lists
graph = [
    [1, 2],
    [0, 2, 3],
    [0, 1, 3],
    [1, 2]
]
print(hamiltonian_paths(graph))
# Time Complexity: O(n!)
# Space Complexity: O(n) due to recursion stack
```

```python
#Word Search II â€” find multiple words (Trie + DFS)
def find_words(board, words):
    R,C=len(board),len(board[0])
    trie={}
    for w in words:
        node=trie
        for ch in w:
            node=node.setdefault(ch,{})
        node['$']=w

    res=[]
    def dfs(i,j,node):
        ch=board[i][j]
        if ch not in node: return
        nxt=node[ch]
        if '$' in nxt:
            res.append(nxt['$']); nxt.pop('$')
        board[i][j]='#'
        for di,dj in ((1,0),(-1,0),(0,1),(0,-1)):
            ni,nj=i+di,j+dj
            if 0<=ni<R and 0<=nj<C and board[ni][nj] != '#':
                dfs(ni,nj,nxt)
        board[i][j]=ch
        if not nxt:
            node.pop(ch)

    for i in range(R):
        for j in range(C):
            dfs(i,j,trie)
    return res
board = [
    ['o','a','a','n'],
    ['e','t','a','e'],
    ['i','h','k','r'],
    ['i','f','l','v']
]
words = ["oath","pea","eat","rain"]
print(find_words(board, words))  # Output: ['oath', 'eat']

# Time Complexity: O(M * N * 4^L) where MxN is board size and L is max word length
# Space Complexity: O(L) due to recursion stack and Trie storage
```

```python
#Different Ways to Add Parentheses (all results)
def diff_ways_to_compute(expr):
    memo={}
    def dfs(s):
        if s.isdigit(): return [int(s)]
        if s in memo: return memo[s]
        res=[]
        for i,ch in enumerate(s):
            if ch in "+-*":
                left=dfs(s[:i]); right=dfs(s[i+1:])
                for a in left:
                    for b in right:
                        if ch=='+': res.append(a+b)
                        elif ch=='-': res.append(a-b)
                        else: res.append(a*b)
        memo[s]=res
        return res
    return dfs(expr)
print(diff_ways_to_compute("2-1-1"))  # Output: [0, 2]
# Time Complexity: O(n * 2^n)
# Space Complexity: O(n) due to recursion stack and memoization
```

```python
#Generate all BSTs with n nodes (construct trees)
class TreeNode:
    def __init__(self,v,left=None,right=None):
        self.val=v; self.left=left; self.right=right

def generate_trees(n):
    from functools import lru_cache
    @lru_cache(None)
    def build(l,r):
        res=[]
        if l>r: return [None]
        for i in range(l,r+1):
            lefts=build(l,i-1); rights=build(i+1,r)
            for L in lefts:
                for R in rights:
                    node=TreeNode(i,L,R)
                    res.append(node)
        return res
    if n==0: return []
    return build(1,n)
trees = generate_trees(3)
print(len(trees))  # Output: 5
# Time Complexity: O(4^n / n^(3/2)) (Catalan number)
# Space Complexity: O(n) due to recursion stack
```

```python
#Kth permutation sequence (recursive selection)
def kth_permutation(n,k):
    import math
    nums=list(range(1,n+1))
    k-=1
    res=[]
    for i in range(n-1,-1,-1):
        f=math.factorial(i)
        idx=k//f
        res.append(str(nums.pop(idx)))
        k%=f
    return "".join(res)
print(kth_permutation(4,9))  # Output: "2314"
# Time Complexity: O(n^2) due to list pop operation
# Space Complexity: O(n) due to recursion stack
```

```python
#All valid multi-type parentheses sequences (e.g., (), [], {})
def multi_paren(n_pairs):
    types = [('(',')'), ('[',']'), ('{','}')]
    res=[]
    opens = [0]*len(types)
    total_pairs = n_pairs
    def dfs(curr, open_count, used_pairs):
        if used_pairs==total_pairs and open_count==0:
            res.append("".join(curr)); return
        # place open if pairs remain
        if used_pairs<total_pairs:
            for i,(o,_) in enumerate(types):
                curr.append(o); opens[i]+=1
                dfs(curr, open_count+1, used_pairs+1)
                opens[i]-=1; curr.pop()
        # place close if matching open exists
        for i,(_,c) in enumerate(types):
            if opens[i]>0:
                curr.append(c); opens[i]-=1
                dfs(curr, open_count-1, used_pairs)
                opens[i]+=1; curr.pop()
    dfs([],0,0)
    return res
# Note: This generates sequences using exactly n_pairs opens total across all types.
print(multi_paren(2))
# Time Complexity: O(k * (2n)!) where k is number of types and
# Space Complexity: O(n) due to recursion stack
```

```python
#Subset sum with reconstruction of all subsets meeting target
def subsets_with_sum(nums, target):
    res=[]
    nums.sort()
    def dfs(i, cur, s):
        if s==target:
            res.append(cur.copy())
            # no return â€” may find more including further zeros
        if i==len(nums): return
        for j in range(i, len(nums)):
            # optional skip duplicates
            cur.append(nums[j])
            dfs(j+1, cur, s+nums[j])
            cur.pop()
    dfs(0, [], 0)
    return res
print(subsets_with_sum([10,1,2,7,6,1,5], 8))
# Time Complexity: O(2^n) in worst case
# Space Complexity: O(n) due to recursion stack
```

```python
#Regular expression matching ('.' and '*') â€” recursive + memo
def is_match(s,p):
    memo={}
    def dfs(i,j):
        if (i,j) in memo: return memo[(i,j)]
        if j==len(p): return i==len(s)
        first_match = i<len(s) and p[j] in {s[i],'.'}
        if j+1<len(p) and p[j+1]=='*':
            ans = dfs(i, j+2) or (first_match and dfs(i+1, j))
        else:
            ans = first_match and dfs(i+1, j+1)
        memo[(i,j)]=ans; return ans
    return dfs(0,0)
print(is_match("aab","c*a*b"))  # Output: True
# Time Complexity: O(m*n) where m and n are lengths of s and p      
# Space Complexity: O(m*n) due to recursion stack and memoization
```

```python
#Generate Gray code sequence (n bits) â€” recursive
def gray_code(n):
    if n==0: return [0]
    prev = gray_code(n-1)
    add = [x | (1<<(n-1)) for x in reversed(prev)]
    return prev + add
print(gray_code(3))  # Output: [0, 1, 3, 2, 6, 7, 5, 4]
# Time Complexity: O(2^n)
# Space Complexity: O(2^n)
```

```python
#Find one Hamiltonian cycle (if exists) â€” backtracking
def find_hamiltonian_cycle(adj):
    n=len(adj)
    path=[0]; visited={0}
    def dfs(u):
        if len(path)==n:
            if 0 in adj[u]: return True
            return False
        for v in adj[u]:
            if v not in visited:
                visited.add(v); path.append(v)
                if dfs(v): return True
                path.pop(); visited.remove(v)
        return False
    if dfs(0):
        return path + [0]
    return None

# adj is adjacency lists; returns cycle path or None
graph = [
    [1, 2],
    [0, 2, 3],
    [0, 1, 3],
    [1, 2]
]   
print(find_hamiltonian_cycle(graph))  # Output: A Hamiltonian cycle path or None
# Time Complexity: O(n!)
# Space Complexity: O(n) due to recursion stack
```

```python
#Combination Sum
def csum(a, target, i=0, path=[]):
    if target==0:
        print(path); return
    if i==len(a) or target<0:
        return
    csum(a,target-a[i],i,path+[a[i]])
    csum(a,target,i+1,path)
csum([2,3,6,7],7)  # Output: [7], [2,2,3]
# Time Complexity: O(2^n) in worst case
# Space Complexity: O(n) due to recursion stack
```

```python
#Combination Sum II (unique)
def csum2(a, target):
    a.sort()
    res=[]
    def dfs(i, t, p):
        if t==0:
            res.append(p); return
        if i==len(a) or t<0:
            return
        dfs(i+1,t-a[i],p+[a[i]])
        j=i
        while j<len(a) and a[j]==a[i]: j+=1
        dfs(j,t,p)
    dfs(0,target,[])
    return res
print(csum2([10,1,2,7,6,1,5],8))  # Output: [[1,1,6],[1,2,5],[1,7],[2,6]]
# Time Complexity: O(2^n) in worst case
# Space Complexity: O(n) due to recursion stack
```

```python
#Generate k-combinations
def k_comb(a, k, i=0, path=[]):
    if len(path)==k:
        print(path); return
    if i==len(a):
        return
    k_comb(a,k,i+1,path+[a[i]])
    k_comb(a,k,i+1,path)
k_comb([1,2,3,4],2)  # Output: [1,2], [1,3], [1,4], [2,3], [2,4], [3,4]
# Time Complexity: O(2^n) in worst case
# Space Complexity: O(n) due to recursion stack
```

```python
#Partition to k Equal Subsets
def k_equal(nums, k):
    s=sum(nums)
    if s%k: return False
    target=s//k
    nums.sort(reverse=True)
    b=[0]*k
    def dfs(i):
        if i==len(nums): return True
        for j in range(k):
            if b[j]+nums[i]<=target:
                b[j]+=nums[i]
                if dfs(i+1): return True
                b[j]-=nums[i]
            if b[j]==0: break
        return False
    return dfs(0)
print(k_equal([4,3,2,3,5,2,1],4))  # Output: True
# Time Complexity: O(k^n) in worst case
# Space Complexity: O(n) due to recursion stack
```

```python
#Solve Knights Tour
def knights(n):
    board=[[ -1]*n for _ in range(n)]
    moves=[(2,1),(1,2),(-1,2),(-2,1),(-2,-1),(-1,-2),(1,-2),(2,-1)]

    def dfs(x,y,m):
        if m==n*n:
            return True
        for dx,dy in moves:
            nx,ny=x+dx,y+dy
            if 0<=nx<n and 0<=ny<n and board[nx][ny]==-1:
                board[nx][ny]=m
                if dfs(nx,ny,m+1): return True
                board[nx][ny]=-1
        return False
    board[0][0]=0
    dfs(0,0,1)
    return board
result = knights(5)
for row in result:
    print(row)
# Time Complexity: O(8^(n^2)) in worst case
# Space Complexity: O(n^2) due to recursion stack and board storage 
```

```python
#Permutations of array of arrays
def perm_arrays(arr, i=0, path=[]):
    if i==len(arr):
        print(path)
        return
    for x in arr[i]:
        perm_arrays(arr, i+1, path+[x])

perm_arrays([[1,2],[3],[4,5]])  # Output: [1,3,4], [1,3,5], [2,3,4], [2,3,5]
# Time Complexity: O(m^n) where m is average length of subarrays and n
# Space Complexity: O(n) due to recursion stack
```

```python
#Generate all possible expressions that evaluate to target
def add_operators(num, target):
    res=[]
    def dfs(i, path, val, last):
        if i==len(num):
            if val==target:
                res.append(path)
            return
        for j in range(i+1, len(num)+1):
            cur = num[i:j]
            if len(cur)>1 and cur[0]=='0': break
            n=int(cur)
            if i==0:
                dfs(j, cur, n, n)
            else:
                dfs(j, path+"+"+cur, val+n, n)
                dfs(j, path+"-"+cur, val-n, -n)
                dfs(j, path+"*"+cur, val-last + last*n, last*n)
    dfs(0,"",0,0)
    return res
print(add_operators("123", 6))  # Output: ["1+2+3", "1*2*3"]
# Time Complexity: O(4^n) in worst case 
# Space Complexity: O(n) due to recursion stack
```

```python
#All paths from source to target in DAG
def all_paths(graph):
    res=[]
    n=len(graph)
    def dfs(u,path):
        if u==n-1:
            res.append(path.copy()); return
        for v in graph[u]:
            dfs(v, path+[v])
    dfs(0,[0])
    return res
graph = [[1,2],[3],[3],[]]
print(all_paths(graph))  # Output: [[0,1,3],[0,2,3]]
# Time Complexity: O(2^n) in worst case
# Space Complexity: O(n) due to recursion stack
```

```python
#Phone keypad (full recursion)
key = {
"2":"abc","3":"def","4":"ghi","5":"jkl",
"6":"mno","7":"pqrs","8":"tuv","9":"wxyz"
}

def keypad(d, s=""):
    if d=="":
        print(s); return
    for c in key[d[0]]:
        keypad(d[1:], s+c)
keypad("23")  # Output: all combinations for digits "23"
# Time Complexity: O(3^n * 4^m) where n is number
# Space Complexity: O(n) due to recursion stack
```

```python
#Decode Parentheses Expression
def decode_par(s):
    def dfs(i):
        res=""
        num=0
        while i < len(s):
            if s[i].isdigit():
                num = num*10 + int(s[i])
            elif s[i]=='[':
                i, sub = dfs(i+1)
                res += sub*num
                num=0
            elif s[i]==']':
                return i, res
            else:
                res += s[i]
            i+=1
        return i, res
    return dfs(0)[1]
print(decode_par("3[a2[c]]"))  # Output: "accaccacc"
# Time Complexity: O(n * k) where k is max repeat count

# Space Complexity: O(n) due to recursion stack and result storage
```
