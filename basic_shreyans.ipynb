```python
#Accept the parameters and calculate the Compound Interest, then print it on STDOUT. Use Python's math module methods.
p = int(input('Enter the principal amount: '))  #p = principal
r = float(input('Enter the rate of interest: '))  #r = rate
t = int(input('Enter the time period: '))      #t = time
ci = p*(pow((1+r/100),t)-1) #ci = compound interest
print('Compound Interest:',ci) #print compound interest
```

```python
#Accept the three sides of a triangle and calculate the area using Heron's formula.
s1 = int(input('Enter the length of side 1: '))
s2 = int(input('Enter the length of side 2: '))
s3 = int(input('Enter the length of side 3: '))
s = (s1+s2+s3)/2 #s = semiperimeter
area = (s*(s-s1)*(s-s2)*(s-s3))**0.5  #formula for Heron's formula is area = sqrt(s(s-a)(s-b)(s-c))
print('Area:',area)
```

```python
#Accept a number and print its reverse.
n = int(input('Enter a number: '))
rev = 0
while n > 0:
    rem = n%10
    rev = rev*10 + rem
    n = n//10
print(rev)
```

```python
'''Accept a number and check if it is a palindrome (If number and its reverse are equal).
Example:
12321 - Reverse - 12321
'''
n = int(input('Enter a number: '))
count=0
while n > 0:
    rem = n%10
    count = count*10 + rem
    n = n//10
if count == n:
    print(n,'is a palindrome')
else:   
    print(n,'is not a palindrome')
    
```
